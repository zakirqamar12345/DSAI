```python
import matplotlib.pyplot as plt

# Data
raisedhands = [15, 20, 10, 30, 40, 42, 35, 50, 12, 70, 50, 19, 5, 20, 62, 30, 36, 55]
visited_resources = [16, 20, 7, 25, 50, 30, 12, 10, 21, 80, 88, 6, 1, 14, 70, 40, 30, 13]
discussion = [20, 25, 30, 35, 50, 70, 17, 22, 50, 70, 80, 12, 11, 19, 60, 66, 80, 90]
target_class = ['M', 'M', 'M', 'M', 'M', 'F', 'M', 'M', 'F', 'F', 'M', 'M', 'M', 'M', 'F', 'F', 'M', 'M']

# Scatter plot
plt.scatter(raisedhands, visited_resources, c=discussion, cmap='viridis', s=80, alpha=0.7)
plt.xlabel("raisedhands")
plt.ylabel("VisITedResources")
plt.title("Scatter Plot with Discussion (Color) and Target Class (Marker)")

# Add text labels for target class
for i in range(len(raisedhands)):
    plt.text(raisedhands[i], visited_resources[i], target_class[i], ha='center', va='center')

plt.colorbar(label="Discussion")
plt.show()
```
