```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file into a DataFrame
#df = pd.read_csv('data.csv')
df = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')
# Define the features
feature1 = 'raisedhands'
feature2 = 'VisITedResources'
feature3 = 'AnnouncementsView'
feature4 = 'Discussion'

# Bar plot
fig, ax = plt.subplots(figsize=(10, 6))
df.groupby('Class')[[feature1, feature2, feature3, feature4]].mean().plot(kind='bar', ax=ax)

plt.xlabel('Class')
plt.ylabel('Mean Value')
plt.title('Mean Values of Features for Each Class')
plt.legend(title='Features')
plt.xticks(rotation=0)
plt.show()
```
