# Linear Regression with gradient descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Extract the features and target variable
X = data[['raisedhands', 'VisITedResources', 'Discussion']]
y = pd.get_dummies(data['Class'])

# Normalize the features
X = (X - X.mean()) / X.std()

# Add a column of ones to X for the intercept term
X.loc[:, 'intercept'] = 1

# Convert y to numpy array
y = y.values

# Initialize theta (weights) with zeros
num_features = X.shape[1]
theta = np.zeros((num_features, y.shape[1]))

# Set hyperparameters
learning_rate = 0.01
iterations = 1000

# Function to perform gradient descent
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for iteration in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradients = (1 / m) * X.T.dot(errors)
        theta = theta - learning_rate * gradients

        cost = (1 / (2 * m)) * np.sum(errors ** 2)
        cost_history.append(cost)

    return theta, cost_history

# Perform gradient descent
theta, cost_history = gradient_descent(X, y, theta, learning_rate, iterations)

# Plot the cost history
plt.plot(range(iterations), cost_history)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost History')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Extract the features and target variable
X = data[['raisedhands', 'Discussion']]
y = pd.get_dummies(data['Class'])

# Normalize the features
X = (X - X.mean()) / X.std()

# Add a column of ones to X for the intercept term
X.loc[:, 'intercept'] = 1

# Convert y to numpy array
y = y.values

# Initialize theta (weights) with zeros
num_features = X.shape[1]
theta = np.zeros((num_features, y.shape[1]))

# Set hyperparameters
learning_rate = 0.01
iterations = 1000

# Function to perform gradient descent
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for iteration in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradients = (1 / m) * X.T.dot(errors)
        theta = theta - learning_rate * gradients

        cost = (1 / (2 * m)) * np.sum(errors ** 2)
        cost_history.append(cost)

    return theta, cost_history

# Perform gradient descent
theta, cost_history = gradient_descent(X, y, theta, learning_rate, iterations)

# Plot the data points
plt.scatter(X['raisedhands'], X['Discussion'], c=np.argmax(y, axis=1))

# Generate grid of values for raisedhands and Discussion
raisedhands_vals = np.linspace(X['raisedhands'].min(), X['raisedhands'].max(), 100)
discussion_vals = np.linspace(X['Discussion'].min(), X['Discussion'].max(), 100)
raisedhands_vals, discussion_vals = np.meshgrid(raisedhands_vals, discussion_vals)

# Compute predicted class for each combination of raisedhands and Discussion
predicted_class = np.argmax(np.dot(np.column_stack((raisedhands_vals.ravel(), discussion_vals.ravel(), np.ones(raisedhands_vals.size))), theta), axis=1)

# Plot the regression line
plt.contourf(raisedhands_vals, discussion_vals, predicted_class.reshape(raisedhands_vals.shape), alpha=0.2, cmap='coolwarm')

plt.xlabel('raisedhands')
plt.ylabel('Discussion')
plt.title('Linear Regression with Gradient Descent')
plt.colorbar(label='Predicted Class')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from matplotlib.colors import ListedColormap

# Read data from CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select relevant features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']

# Encode the target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Create a DataFrame with feature names
feature_names = features.columns.tolist()
features = pd.DataFrame(features, columns=feature_names)

# Create an instance of the Decision Tree classifier
dt = DecisionTreeClassifier()

# Fit the classifier to the data
dt.fit(features, target)

# Define the colormap for the plot
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Create a meshgrid of feature values
x_min, x_max = features['raisedhands'].min() - 1, features['raisedhands'].max() + 1
y_min, y_max = features['VisITedResources'].min() - 1, features['VisITedResources'].max() + 1
z_min, z_max = features['Discussion'].min() - 1, features['Discussion'].max() + 1
xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, 1.0),
                         np.arange(y_min, y_max, 1.0),
                         np.arange(z_min, z_max, 1.0))

# Predict the target variable for each point in the meshgrid
Z = dt.predict(np.c_[xx.ravel(), yy.ravel(), zz.ravel()])

# Reshape the predictions to match the meshgrid shape
Z = Z.reshape(xx.shape)

# Create a contour plot of the decision boundaries
plt.figure(figsize=(10, 8))
plt.contourf(xx[:, :, 0], yy[:, :, 0], Z[:, :, 0], alpha=0.4, cmap=cmap)
plt.colorbar()

# Plot the data points
plt.scatter(features['raisedhands'], features['VisITedResources'], features['Discussion'], c=target, cmap=cmap)

plt.xlabel('raisedhands')
plt.ylabel('VisITedResources')
plt.title('Decision Tree Decision Boundaries')

plt.show()
```

# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select the features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']  # Replace 'target_variable' with the actual target column name

# Encode the non-numeric target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Create an instance of the Linear Regression model
linreg = LinearRegression()

# Fit the model to the data
linreg.fit(features, target)

# Generate points for the regression plane
x_surf, y_surf = np.meshgrid(features['raisedhands'], features['VisITedResources'])
z_surf = linreg.predict(pd.DataFrame({'raisedhands': x_surf.ravel(), 'VisITedResources': y_surf.ravel(), 'Discussion': np.zeros_like(x_surf).ravel()})).reshape(x_surf.shape)

# Plot the scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features['raisedhands'], features['VisITedResources'], target)

# Plot the regression plane
ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, cmap='viridis')
ax.set_xlabel('raisedhands')
ax.set_ylabel('VisITedResources')
ax.set_zlabel('target_variable')
plt.show()
```
