# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select the features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']  # Replace 'target_variable' with the actual target column name

# Encode the non-numeric target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create an instance of the Linear Regression model
linreg = LinearRegression()

# Fit the model to the training data
linreg.fit(X_train, y_train)

# Predict the target variable for the testing set
y_pred = linreg.predict(X_test)

# Calculate the coefficient of determination (R-squared score)
accuracy = r2_score(y_test, y_pred)
print("R-squared score:", accuracy)

# Generate points for the regression plane
x_surf, y_surf = np.meshgrid(features['raisedhands'], features['VisITedResources'])
z_surf = linreg.predict(pd.DataFrame({'raisedhands': x_surf.ravel(), 'VisITedResources': y_surf.ravel(), 'Discussion': np.zeros_like(x_surf).ravel()})).reshape(x_surf.shape)

# Plot the scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features['raisedhands'], features['VisITedResources'], target)

# Plot the regression plane
ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, cmap='viridis')
ax.set_xlabel('raisedhands')
ax.set_ylabel('VisITedResources')
ax.set_zlabel('target_variable')
plt.show()
```

# Logistic Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Read data from CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select relevant features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']

# Encode the target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create an instance of the LogisticRegression classifier
logreg = LogisticRegression()

# Fit the classifier to the training data
logreg.fit(X_train, y_train)

# Predict the target variable for the testing set
y_pred = logreg.predict(X_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Create a scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features['raisedhands'], features['VisITedResources'], features['Discussion'], c=target, cmap='viridis')

ax.set_xlabel('raisedhands')
ax.set_ylabel('VisITedResources')
ax.set_zlabel('Discussion')

plt.show()
```

# Neural Network

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# Read data from CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select relevant features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']

# Encode the target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Create an instance of the MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=1000)

# Perform cross-validation and calculate accuracy scores
accuracy_scores = cross_val_score(mlp, features, target, cv=5)

# Calculate the mean accuracy score
mean_accuracy = accuracy_scores.mean()

# Print the accuracy scores and mean accuracy
print("Accuracy Scores:", accuracy_scores)
print("Mean Accuracy:", mean_accuracy)

# Create a scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features['raisedhands'], features['VisITedResources'], features['Discussion'], c=target, cmap='viridis')

ax.set_xlabel('raisedhands')
ax.set_ylabel('VisITedResources')
ax.set_zlabel('Discussion')

plt.show()
```

# Decision Tree

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# Read data from CSV file
data = pd.read_csv(r'E:\NUST-dairy\Module\Project\AI-Data.csv')

# Select relevant features and target variable
features = data[['raisedhands', 'VisITedResources', 'Discussion']]
target = data['Class']

# Encode the target variable
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target)

# Create an instance of the Decision Tree classifier
dt = DecisionTreeClassifier()

# Perform cross-validation and calculate accuracy scores
accuracy_scores = cross_val_score(dt, features, target, cv=5)

# Calculate the mean accuracy score
mean_accuracy = accuracy_scores.mean()

# Print the accuracy scores and mean accuracy
print("Accuracy Scores:", accuracy_scores)
print("Mean Accuracy:", mean_accuracy)

# Create a scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(features['raisedhands'], features['VisITedResources'], features['Discussion'], c=target, cmap='viridis')

ax.set_xlabel('raisedhands')
ax.set_ylabel('VisITedResources')
ax.set_zlabel('Discussion')

plt.show()
```
